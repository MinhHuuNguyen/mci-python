## 1. Overall introduction
Matplotlib is an open-source Python library for data visualization. <br>
Matplotlib is the whole package. And it contains 2 modules: `pylab` and `pyplot`. <br>
- `pyplot` provides the state-machine interface for plotting. That means, each line of code change the state of plotting.
- `pylab` can be said as the combination of `numpy` functionality and `pyplot` functionality.

## 3. Simple plots
### 3.1. Line plot
#### Show single line plot and multiple line plots

```python
x = np.linspace(-5, 5, 50)
```

```python
# single line plot
plt.plot(x, np.sin(x))
plt.show()
```

```python
# multiple line plots in multiple lines of code
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# multiple lines plot in single line of code
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

#### Change Marker format
##### Change marker shape by `marker`
List of marker shapes [here](https://matplotlib.org/stable/api/markers_api.html)

```python
plt.plot(x, np.sin(x - 0), marker='p')
plt.plot(x, np.sin(x - 2), marker='v')
plt.plot(x, np.sin(x - 4), marker='<')
plt.show()
```

##### Change marker size by `markersize` or `ms`

```python
# Using `markersize`
plt.plot(x, np.sin(x - 0), marker='o', markersize=3)
plt.plot(x, np.sin(x - 2), marker='o', markersize=6)
plt.plot(x, np.sin(x - 4), marker='o', markersize=9)
plt.show()
```

```python
# Using `ms`
plt.plot(x, np.sin(x - 0), marker='o', ms=3)
plt.plot(x, np.sin(x - 2), marker='o', ms=6)
plt.plot(x, np.sin(x - 4), marker='o', ms=9)
plt.show()
```

##### Change marker edge color by `markeredgecolor` or `mec`

```python
# Using `markeredgecolor`
plt.plot(x, np.sin(x - 0), marker='o', markeredgecolor='r')
plt.plot(x, np.sin(x - 2), marker='o', markeredgecolor='g')
plt.plot(x, np.sin(x - 4), marker='o', markeredgecolor='b')
plt.show()
```

```python
# Using `mec`
plt.plot(x, np.sin(x - 0), marker='o', mec='r')
plt.plot(x, np.sin(x - 2), marker='o', mec='g')
plt.plot(x, np.sin(x - 4), marker='o', mec='b')
plt.show()
```

##### Change marker face color by `markerfacecolor` or `mfc`

```python
# Using `markerfacecolor`
plt.plot(x, np.sin(x - 0), marker='o', markerfacecolor='r')
plt.plot(x, np.sin(x - 2), marker='o', markerfacecolor='g')
plt.plot(x, np.sin(x - 4), marker='o', markerfacecolor='b')
plt.show()
```

```python
# Using `mfc`
plt.plot(x, np.sin(x - 0), marker='o', mfc='r')
plt.plot(x, np.sin(x - 2), marker='o', mfc='g')
plt.plot(x, np.sin(x - 4), marker='o', mfc='b')
plt.show()
```

#### Change line color by `color` or `c`
List of colors [here](https://matplotlib.org/stable/tutorials/colors/colors.html)

```python
# Using `color`
plt.plot(x, np.sin(x - 0), color='blue')        # Color by name
plt.plot(x, np.sin(x - 1), color='g')           # Color by shortname
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.show()
```

```python
# Using `c`
plt.plot(x, np.sin(x - 0), c='blue')        # Color by name
plt.plot(x, np.sin(x - 1), c='g')           # Color by shortname
plt.plot(x, np.sin(x - 2), c='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), c='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), c=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.show()
```

#### Change line style by `linestyle` or `ls`

```python
# Using `linestyle`
plt.plot(x, x - 1, linestyle='None')
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

plt.plot(x, x - 1, linestyle='')
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':')  # dotted
plt.show()
```

```python
# Using `ls`
plt.plot(x, x - 1, ls='None')
plt.plot(x, x + 0, ls='solid')
plt.plot(x, x + 1, ls='dashed')
plt.plot(x, x + 2, ls='dashdot')
plt.plot(x, x + 3, ls='dotted')

plt.plot(x, x - 1, ls='')   # None
plt.plot(x, x + 4, ls='-')  # solid
plt.plot(x, x + 5, ls='--') # dashed
plt.plot(x, x + 6, ls='-.') # dashdot
plt.plot(x, x + 7, ls=':')  # dotted
```

#### Change line width by `linewidth` or `lw`

```python
# Using `linewidth`
plt.plot(x, x - 1, linewidth='3')
plt.plot(x, x + 2, linewidth='6')
plt.plot(x, x + 5, linewidth='9')
plt.show()
```

```python
# Using `lw`
plt.plot(x, x - 1, lw='3')
plt.plot(x, x + 2, lw='6')
plt.plot(x, x + 5, lw='9')
plt.show()
```

#### Change line format by Format String
Format String = `'[marker][line][color]'`

```python
plt.plot(x, x + 0, '-g')    # [no marker][solid][green]
plt.plot(x, x + 2, 'o--c')  # [circle][dashed][cyan] 
plt.plot(x, x + 4, 's-.k')  # [square][dashdot][black]
plt.show()
```

### 3.2. Scatter plot

List of color map style [here](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
color_map = np.arange(50)

plt.scatter(x, y, c=color_map, cmap='viridis')
plt.colorbar()
plt.show()
```

Bubble plot is Scatter plot with multiple sizes of scatter

```python
color_map = np.arange(50)
bubble_size = np.random.randint(1, 200, size=50)

plt.scatter(x, y, s=bubble_size, c=color_map, cmap='viridis')
plt.colorbar()
plt.show()
```

## 6. Subplots - Multiple plots
### 6.1. Create single sub-plot by `plt.subplot()`

```python
for subplot_index in range(4):
    plt.subplot(3, 2, subplot_index+1)
    plt.text(0.5, 0.5, f'subplot_index_{subplot_index}', fontsize=12, horizontalalignment='center')
# Add space between subplots
plt.tight_layout()
plt.show()
```

### 6.2. Create the whole sub-plots by `plt.subplots()`

```python
num_rows, num_cols = 3, 2
fig, ax = plt.subplots(num_rows, num_cols)

for row_index in range(num_rows):
    for col_index in range(num_cols):
        ax[row_index, col_index].text(
            0.5, 0.5, f'subplot_index_{row_index}_{col_index}', fontsize=12, horizontalalignment='center')

# Add space between subplots
plt.tight_layout()
plt.show()
```

### 6.3. More specific subplots by using `plt.GridSpec()`

```python
grid = plt.GridSpec(3, 3)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1:, :2])
plt.subplot(grid[1:, 2])
# Add space between subplots
plt.tight_layout()
plt.show()
```

### 6.4. Multiple plots in one figure by `twinx()`

```python
x = np.linspace(-5, 5, 50)
```

```python
fig, ax_1 = plt.subplots()

ax_1.plot(x, np.sin(x), label='sin(x)')
ax_1.set_ylabel('sin(x)')

ax_2 = ax_1.twinx()
ax_2.plot(x, np.exp(x), 'r.-', label='exp(x)')
ax_2.set_ylabel('exp(x)')

fig.legend(loc='upper left')
plt.show()
```

```python
x = np.arange(1, 11)
y = np.random.randint(0, 100, 10)
z = np.random.randint(0, 100, 10)
```

```python
fig, ax_1 = plt.subplots()

ax_1.bar(x, y, label='y_value')
ax_1.set_ylabel('y_value')

ax_2 = ax_1.twinx()
ax_2.plot(x, z, 'r.-', label='z_value')
ax_2.set_ylabel('z_value')

fig.legend(loc='upper left')
plt.show()
```
