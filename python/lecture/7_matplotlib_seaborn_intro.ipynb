## 4. Line plot

#### Show single line plot

```python
plt.plot(x, np.sin(x))
```

#### Show multiple lines plot

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

#### Change line color

```python
plt.plot(x, np.sin(x - 0), color='blue')        # Color by name
plt.plot(x, np.sin(x - 1), color='g')           # Color by 
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

#### Change line style

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

#### Change both line color and line style

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## 5. Axes limit

#### Limit axes by `plt.xlim()` and `plt.ylim()`

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

#### Reverse the axis

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

#### Limit axes by `plt.axis()`

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

#### Automatically limit axis

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

#### Ensure aspect ratio

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## 6. Plot title, label and legends

```python
plt.plot(x, np.sin(x), label='sin(x)') # Put label here to show in legend
plt.plot(x, np.cos(x), label='cos(x)') # Put label here to show in legend
plt.title("sin(x) vs cos(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
```

## 9. Sub-plot

### Create single sub-plot by `plt.subplot()`

```python
for subplot_index in range(1, 10):
    plt.subplot(3, 3, subplot_index)
    plt.text(0.5, 0.5, f'subplot_index_{subplot_index}', fontsize=12, ha='center')
```

### Create the whole sub-plots by `plt.subplots()`

```python
num_rows, num_cols = 3, 3
fig, ax = plt.subplots(num_rows, num_cols)

for row_index in range(num_rows):
    for col_index in range(num_cols):
        ax[row_index, col_index].text(
            0.5, 0.5, f'subplot_index_{row_index}_{col_index}', fontsize=12, ha='center')

# Add space between subplots
fig.tight_layout(pad=1.0)
```

### More specific subplots by using `plt.GridSpec()`

```python
grid = plt.GridSpec(3, 3, wspace=0.4, hspace=0.3)

plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1:, :2])
plt.subplot(grid[1:, 2]);
```
