## Homework
### Exercise 1:
Create dummy dataset which contains 5 data fields: "first_name", "last_name", "age", "job", "country" and 10 rows. (dataset contains 1 "country" named "Vietnam", 1 age is "20")

### Exercise 2:
Create 5 Series for each field, after that, create a Dataframe from 5 Series, dump that Dataframe to csv file.

### Exercise 3:
Read the csv file from excercise 2, add the string "my_first_name" in to each value in the "first_name", add the string "my_last_name" in to each value in the "last_name".

Print all rows whose "country" is "Vietnam", whose "country" is not "Vietnam", whose "age" is greater than "20", whose "age" is lower than or equal "20".

## Solution

```python
import pandas as pd
```

### Excercise 1:

```python
first_name_list = [
    'David',
    'Elon',
    'Ho Chi',
    'Robert',
    'Michael',
    'Tony',
    'Pham Nhat',
    'Truong Gia',
    'Jack',
    'Leonardo'
]
```

```python
last_name_list = [
    'Beckham',
    'Musk',
    'Minh',
    'Downey Jr.',
    'Nguyen',
    'Tam',
    'Vuong',
    'Binh',
    'Ma',
    'DiCaprio'
]
```

```python
age_list = [
    46,
    50,
    79,
    56,
    20,
    10,
    53,
    65,
    57,
    47
]
```

```python
job_list = [
    'footballer',
    'CEO',
    'president',
    'actor',
    'data scientist',
    'student',
    'businessman',
    'businessman',
    'businessman',
    'actor'
]
```

```python
country_list = [
    'England',
    'South Africa',
    'Vietnam',
    'America',
    'England',
    'America',
    'Vietnam',
    'Vietnam',
    'China',
    'America'
]
```

### Excercise 2:

```python
first_name_series = pd.Series(first_name_list, name='first_name')
first_name_series
```

```python
last_name_series = pd.Series(last_name_list, name='last_name')
last_name_series
```

```python
age_series = pd.Series(age_list, name='age')
age_series
```

```python
job_series = pd.Series(job_list, name='job')
job_series
```

```python
country_series = pd.Series(country_list, name='country')
country_series
```

```python
info_df = pd.concat(
    [first_name_series, last_name_series, age_series, job_series, country_series],
    axis=1
)
info_df
```

```python
info_df.to_csv('info.csv', index=False)
```

### Excercise 3:

```python
new_info_df = pd.read_csv('info.csv')
new_info_df
```

```python
new_info_df.first_name = new_info_df.first_name.apply(lambda x: f'my_first_name_{x}')
```

```python
new_info_df
```

```python
new_info_df.last_name = new_info_df.last_name.apply(lambda x: f'{x}_my_last_name')
```

```python
new_info_df
```

### Excercise 4:

```python
new_info_df.age = new_info_df.age.apply(int)
```

```python
new_info_df.country = new_info_df.country.apply(lambda x: x.lower())
```

```python
new_info_df
```

```python
new_info_df.loc[new_info_df.country == 'vietnam']
```

```python
new_info_df.loc[new_info_df.country != 'vietnam']
```

```python
new_info_df.loc[new_info_df.age <= 20]
```

```python
new_info_df.loc[new_info_df.age > 20]
```
