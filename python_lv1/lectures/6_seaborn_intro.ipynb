## 3. Using `matplotlib` with and without `seaborn`
### 3.1. Use `matplotlib` only

```python
import numpy as np
from matplotlib import pyplot as plt

def sin_plot():
    x = np.linspace(-10, 10, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x - i))

sin_plot()
plt.show()
```

### 3.2. Use `matplotlib` with `seaborn`

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

def sin_plot():
    x = np.linspace(-10, 10, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x - i))

sin_plot()
plt.show()
```

### 3.3. Some figure styles in `seaborn`

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
# white, dark, whitegrid, darkgrid, ticks
sns.set_style('darkgrid')

def sin_plot():
    x = np.linspace(-10, 10, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x - i))

sin_plot()
plt.show()
```

### 3.4. `matplotlib` and `seaborn` can be used together

```python
x = np.linspace(-10, 10, 100)
data = pd.DataFrame(data={
    'x': x,
    'sin_x': np.sin(x)
})
```

```python
sns.set()
```

```python
# seaborn with plt.title()
sns.lineplot(x='x', y='sin_x', data=data)
plt.title('x and sin(x)')
plt.show()
```

```python
# seaborn with plt.xlim() and plt.ylim()
sns.lineplot(x='x', y='sin_x', data=data)
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.show()
```

```python
# seaborn with plt.subplot()
for subplot_index in range(6):
    plt.subplot(3, 2, subplot_index+1)
    sns.lineplot(x='x', y='sin_x', data=data)

plt.tight_layout()
plt.show()
```

## 5. Simple plots
### 5.1. Line plot
#### Line plot with single category

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=data_iris)
plt.show()
```

#### Line plot with multiple categories

```python
sns.lineplot(x='sepal_length', y='sepal_width', hue='species', data=data_iris)
plt.show()
```

### 5.2. Scatter plot
#### Scatter plot with single category

```python
sns.scatterplot(x='petal_length', y='petal_width', data=data_iris)
plt.show()
```

#### Scatter plot customize marker

```python
sns.scatterplot(x='petal_length', y='petal_width', data=data_iris, marker='^')
plt.show()
```

#### Scatter plot with multiple categories

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=data_iris)
plt.show()
```

#### Scatter plot with single category with distributions

```python
sns.jointplot(x='petal_length', y='petal_width', data=data_iris)
plt.show()
```

#### Scatter plot with multiple categories with distributions

```python
sns.jointplot(x='petal_length', y='petal_width', hue='species', data=data_iris)
plt.show()
```

#### Scatter plot with regression curve

```python
sns.regplot(x='sepal_length', y='petal_length', data=data_iris)
plt.show()
```

#### Scatter plot with KDE and regression curve

```python
sns.jointplot(x='petal_length', y='petal_width', data=data_iris, kind='reg')
plt.show()
```

### 5.3. Bar plot

#### Bar plot with multiple categories

### 5.4. Histogram
#### Histogram with `kde`
`kde` is `Kernel Density Estimation`

```python
# kde=False
sns.histplot(x='petal_length', kde=False, data=data_iris)
plt.show()
```

```python
# kde=True
sns.histplot(x='petal_length', kde=True, data=data_iris)
plt.show()
```

```python
# kde=True
x = np.random.normal(100, size=100000)
sns.histplot(x, kde=True, bins=100)
plt.show()
```

#### Histogram with multiple categories by `hue`

```python
sns.histplot(x='petal_length', kde=True, data=data_iris)
plt.show()
```

```python
sns.histplot(x='petal_length', hue='species', kde=True, data=data_iris)
plt.show()
```

### 6.2. Box plot
or `box and whisker plot`

<img src="../images/sb_box_plot.png" width="400px"/>

### 6.3. Violin plot
#### Mapping between Box plot and Violin plot
<img src="../images/sb_box_violin.png" width="350px"/>

#### Why Violin plot is better than Box plot?
<img src="../images/sb_box_violin.gif" width="600px"/>

#### Violin plot with single category

```python
# Box plot
sns.boxplot(x='class', y='age', data=data_titanic)
plt.show()
```

```python
# Violin plot
sns.violinplot(x='class', y='age', data=data_titanic)
plt.show()
```

#### Violin plot with multiple categories

```python
# Box plot
sns.boxplot(x='class', y='age', hue='sex', data=data_titanic)
plt.show()
```

```python
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=data_titanic)
plt.show()
```

```python
# Violin plot with split=True
sns.violinplot(x='class', y='age', hue='sex', data=data_titanic, split=True)
plt.show()
```

#### With more than two categories, we cannot use `split=True`

```python
sns.boxplot(x='sex', y='age', hue='class', data=data_titanic)
plt.show()
```

```python
sns.violinplot(x='sex', y='age', hue='class', data=data_titanic)
plt.show()
```

```python
sns.violinplot(x='sex', y='age', hue='class', data=data_titanic, split=True)
plt.show()
```
