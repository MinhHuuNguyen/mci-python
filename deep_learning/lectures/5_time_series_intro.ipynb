````python
from datetime import datetime

import yfinance as yf
import pandas as pd
import numpy as np
````

#### Prepare data

````python
tech_list = ['AAPL', 'GOOG', 'MSFT', 'AMZN']
````

````python
end = datetime.now()
start = datetime(end.year - 30, end.month, end.day)
````

````python
df_list = []
for stock in tech_list:
    df = yf.download(stock, start, end)
    df = df.reset_index()
    df['company'] = stock
    df_list.append(df)
````

````python
df = pd.concat(df_list)
df
````
